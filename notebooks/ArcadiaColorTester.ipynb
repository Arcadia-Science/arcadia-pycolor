```python
def randspline_colortest(dictionary, indexes, timepoints = 6, title = '', save = ''):
    color_list = [list(dictionary.keys())[i] for i in indexes]
    
    for color in color_list:
        x = np.arange(timepoints)
        y = np.random.rand(timepoints)
        
        spline = make_interp_spline(x, y)
        
        X_ = np.linspace(x.min(), x.max(), 500)
        Y_ = spline(X_)
        
        plt.plot(X_, Y_, label = color, color = dictionary[color], linewidth = 3)
    
    plt.title(title)
    plt.legend(loc = 'upper right', bbox_to_anchor = (1.4, 1))
    if save != '':
        plt.savefig(save)
    plt.show()
```

```python
randspline_colortest(
    dictionary = accent_v3,
    indexes = list(np.arange(0, 6)) + list(np.arange(12, 14)),
    title = '8 Bold Color Test',
    save = '8boldcolortest.pdf'
)
randspline_colortest(
    dictionary = accent_v3,
    indexes = list(np.arange(0, 6)) + list(np.arange(12, 16)),
    title = '10 Bold Color Test',
    save = '10boldcolortest.pdf'
)
randspline_colortest(
    dictionary = accent_v3,
    indexes = list(np.arange(0, 6)) + list(np.arange(12, 18)),
    title = '12 Bold Color Test',
    save = '12boldcolortest.pdf'
)
```

```python
def plot_examples(colormaps):
    """
    Helper function to plot data with associated colormap.
    """
    np.random.seed(19680801)
    data = np.random.randn(30, 30)
    n = len(colormaps)
    fig, axs = plt.subplots(1, n, figsize=(n * 2 + 2, 3),
                            constrained_layout=True, squeeze=False)
    for [ax, cmap] in zip(axs.flat, colormaps):
        psm = ax.pcolormesh(data, cmap=cmap, rasterized=True, vmin=-4, vmax=4)
        fig.colorbar(psm, ax=ax)
    plt.show()
```

```python
arcadia_color_rgb = mpl.colors.to_rgba('arcadia:amber')
N = 256
vals = np.ones((N, 4))
vals[:, 0] = np.linspace(arcadia_color_rgb[0], 1, N)
vals[:, 1] = np.linspace(arcadia_color_rgb[1], 1, N)
vals[:, 2] = np.linspace(arcadia_color_rgb[2], 1, N)
newcmp = ListedColormap(vals)
viridis = mpl.colormaps['viridis']
plot_examples([viridis, newcmp])
```

```python
def make_gradient_cmap(from_color, to_color, steps, name):
    import numpy as np
    import matplotlib as mpl
    
    if type(from_color) == str:
        from_color_rgb = mpl.colors.to_rgba(from_color)
    if type(to_color) == str:
        to_color_rgb = mpl.colors.to_rgba(to_color)
    
    vals = np.ones((steps, 4))
    vals[:, 0] = np.linspace(from_color_rgb[0], to_color_rgb[0], N)
    vals[:, 1] = np.linspace(from_color_rgb[1], to_color_rgb[1], N)
    vals[:, 2] = np.linspace(from_color_rgb[2], to_color_rgb[2], N)
    
    return ListedColormap(vals, name = name)
```

```python
aegean_cmap = make_gradient_cmap('arcadia:aegean', 'white', 256, 'aegean')

viridis = mpl.colormaps['viridis']
plot_examples([viridis, aegean_cmap])
```

```python
aegean_cmap_r = aegean_cmap.reversed()
plot_examples([aegean_cmap, aegean_cmap_r])
```

```python
mpl.colormaps.register(cmap=aegean_cmap)
mpl.colormaps.register(cmap=aegean_cmap_r)

data = [[1, 2, 3, 4, 5]]

fig, (ax1, ax2) = plt.subplots(nrows=2)

ax1.imshow(data, cmap='aegean')
ax2.imshow(data, cmap='aegean_r')

plt.show()
```
